Reading Datasets:
## data 1 = asd_children_traits
## data 2 = adults 
## data 3 = toddler autism dataset 

```python
import pandas as pd
```

```python


data1=pd.read_csv('DATA/asd_children_traits.csv')
data2=pd.read_csv('DATA/Toddler Autism dataset July 2018.csv')
data3=pd.read_csv('DATA/autism_screening_adults.csv')

```

```python
data2.head()
```

```python
data3.head()
```

```python
print(data1.columns)
print(data2.columns)
print(data3.columns)
```

```python
df1=pd.concat([data1.iloc[:,1:11],data1.iloc[:,[12,22,23,24,25,26,27]]],axis=1)
df1.head()
```

```python
df2=pd.concat([data2.iloc[:,1:12],data2.iloc[:,13:]],axis=1)
df2['Age_Mons']=(df2['Age_Mons']/12).astype(int)
df2.head()
```

```python
df3=pd.concat([data3.iloc[:,0:15],data3.iloc[:,-2:]],axis=1)
df3.head()
```

```python
order_test= pd.DataFrame({
    'df1': df1.columns,
    'df2': df2.columns ,
    'df3': df3.columns
})
order_test

```

```python
# Rename columns to have the same names in all DataFrames
df2.columns = df3.columns = df1.columns

# Concatenate the DataFrames
data_fin = pd.concat([df3, df2, df1], axis=0)
data_fin.head()

```

```python
data_fin.shape
```

```python
# Get object type columns
object_cols = data_fin.select_dtypes('O').columns

# Create new DataFrame
object_df = pd.DataFrame({
    'Objects': object_cols,
    'Unique values': [data_fin[col].unique() for col in object_cols],
    'number of unique values':[data_fin[col].nunique()for col in object_cols]
})

object_df
```

```python
data_fin.columns


for col in ['Sex', 'Ethnicity',
       'Jaundice', 'Family_mem_with_ASD', 'Who_completed_the_test',
       'ASD_traits']:
    print("-------------------------------")
    print(f'Column name: {col}\n')
    print(f'Unique values:\n{data_fin[col].unique()}')

```

```python
replacements = {
    'f': 'F',
    'm': 'M',
}
data_fin['Sex'] = data_fin['Sex'].replace(replacements)
```

```python
replacements = {
    'yes': 'Yes',
    'no': 'No',
}
data_fin['Jaundice'] = data_fin['Jaundice'].replace(replacements)
```

```python
replacements = {
    'yes': 'Yes',
    'no': 'No',
}
data_fin['Family_mem_with_ASD'] = data_fin['Family_mem_with_ASD'].replace(replacements)

```

```python
replacements = {
    'YES': 'Yes',
    'NO': 'No',
}
data_fin['ASD_traits'] = data_fin['ASD_traits'].replace(replacements)
```

```python
replacements = {
    'middle eastern': 'Middle Eastern',
    'Middle Eastern ': 'Middle Eastern',
    'mixed': 'Mixed',
    'asian': 'Asian',
    'black': 'Black',
    'south asian': 'South Asian',
    'PaciFica':'Pacifica',
    'Pasifika':'Pacifica'
    
}
data_fin['Ethnicity'] = data_fin['Ethnicity'].replace(replacements)
```

```python
replacements = {
    'Health care professional':'Health Care Professional',
    'family member':'Family Member',
    'Family member':'Family Member'
}
data_fin['Who_completed_the_test'] = data_fin['Who_completed_the_test'].replace(replacements)
```

```python
for col in ['Sex', 'Ethnicity',
       'Jaundice', 'Family_mem_with_ASD', 'Who_completed_the_test',
       'ASD_traits']:
    print("-------------------------------")
    print(f'Column name: {col}\n')
    print(f'Unique values:\n{data_fin[col].unique()}')

```

treating other missing values as null 

```python
# Find the mode of the 'Ethnicity' column
mode_ethnicity = data_fin['Ethnicity'].mode()[0]
#mode_ethnicity

data_fin['Ethnicity'].fillna(mode_ethnicity, inplace=True)
```

```python


# Find the mode of the 'Who_completed_the_test' column
mode = data_fin['Who_completed_the_test'].mode()[0]

data_fin['Who_completed_the_test'].fillna(mode, inplace=True)

```

```python


# Find the mode of the 'Age_Years' column
mode = data_fin['Age_Years'].mode()[0]

data_fin['Age_Years'].fillna(mode, inplace=True)

```
